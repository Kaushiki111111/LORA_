#Generate a rank deficient matrix:

```python
#Lets say we have a matrix with dim=25 and we keep rank as 5
```

```python
dim=10
k=10
r=2
```

```python
W=torch.randn(dim,r) @ torch.randn(r,k)
```

```python
W
```

#evaluate rank of matrix:

```python
W_rank=np.linalg.matrix_rank(W)
print(f'Rank of W is :{W_rank}')
```

#Now lets compute the A nd B ,SVD rank decomposition matrices:

Perform svd on W=UAV^T

```python
U, A, Vh = torch.linalg.svd(W)
```

```python
U
```

```python
A
```

```python
Vh

```

For rank r factorization keep only r singular values:

```python
U_r= U[:,:W_rank]
A_r =torch.diag(A[: W_rank])
Vh_r= Vh[:,: W_rank].t()
```

```python
A_r
```

```python
A_m=U_r @ A_r
```

```python
B=Vh_r
```

```python
print(f'Shape of A matrix: {A_m.shape}')
```

```python
print(f'Shape of B matrix: {B.shape}')
```
